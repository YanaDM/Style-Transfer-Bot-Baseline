# Финальный проект первого семестра DLSchool. 

## Style transfer telegram bot.

В данном проекте мы хотим сделать телеграм бота, который по двум картинкам будет переносить стиль с одной на другую.

В данной проекте у нас есть две папки:
* model - папка для ноутбуков с экспериментами. В этой папаке вы можете создавать ноутбуки и тестировать работоспособность моделей, которые вы нашли.
* telegram_bot - папка с .py файлами, которые относятся к продакшн части нашего проекта. Здесь будет лежать наш файл с телеграм ботом, а так же модель, выполняющая пернеос стиля.

Основной целью данного проекта является погружение в тему Style Transfer, изучение нескольких методик и применение готовых решений в боте (хотя Вы можете напистаь свою реализацию алгоритма из статей). Вначале Вы поработаете с алгоритмом быстрого переноса стиля от Johnson et al, а потом и с другими.

В бейзлайне используется библиотека для ботов python telegram bot. Документация к ней легко гуглится. 

Если при запуске бейзлайна вам не хватает какой-то библиотеки - просто забивайте в гугл название отсутствующего модуля и pip'ом устанавливайте нужную библиотеку.

Думаю, что наша работа будет максимально продуктивной, если вы будете двигаться по следющему курсу:

0. Написать BotFather и получить свой токен. В папке telegram_bot создать файл telegram_token.py с единственной строчкой: token = 'ВАШ ТОКЕН ЗДЕСЬ'. **Будьте осторожны и не загрузите свой токен на гитхаб!!!** Я уже включил файл telegram_token.py в .gitignore, поэтому токен не попадет на гитхаб, если вы будете держать его только в этом файле.
1. Созддайте свой гитхаб репозиторий и сделайте начальный коммит, состоящий из файлов бейзлайна, и скиньте ссылку в слак, чтобы я имел доступ к репозиторию. Эту тетрадку включать не стоит, для удобства я внесу её в .gitignore. Если вы не умеете в гит, то напишите мне, я помогу разобраться.
2. Изучите статью и заметки об алгоритме переноса стиля: https://arxiv.org/pdf/1603.08155.pdf, https://towardsdatascience.com/perceptual-losses-for-real-time-style-transfer-and-super-resolution-637b5d93fa6d. Эти знания понадобятся, чтобы аккуратно и с пониманием перенести готовую модель из репозитория https://github.com/lengstrom/fast-style-transfer (эта реализация основана на несокльких статьях и использует очень важное нововведение по сравнению с johnson et al - Instance Normalisaton вместо BatchNorm, статья про это https://arxiv.org/abs/1607.08022). Добавьте несколько разных предобученных стилей и дайте пользователю выбор. 
3. Доработайте самого бота, чтобы с ним было приятно общаться: добавьте приветствие и объяснение принципа работы, сделайте кнопки, чтобы человек мог переходить в режим стайл трансфера и выходить из него. Вынесите все константы в конфиг.
4. На этом шаге Вы можете обучить один или несколько своих стилей и также добавить их в бота.
5. Теперь можно добавить поддержку других алгоритмов для переноса стиля.
  * Алгоритмы, основанные на GAN. Их преимущество в том, что они работают быстро и позволяют переносить любой стиль.
  * Алгоритмы фотореалистичного переноса стиля. Такие алгоритмы позволяют переносить стиль, оставляя картинку похожей на фото. Например, таким алгоритмом можно сменить время года на картинке.
  * итд
  
  Перед тем как начать работу над добавлением поддержки другого алгоритма, напишите мне, и я помогу Вам определиться с направлением и найти материалы. 
6. Теперь все готово и остались только косметические изменения: добавление docstring, исправение кодстайла, рефакторинг и т.д. 
7. ...
8. PROFIT!!!
